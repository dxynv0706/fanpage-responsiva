```python
import matplotlib.pyplot as plt

class Biblioteca:
    def __init__(self):
        self.livros = []

    # Criar e cadastrar livro
    def cadastrar_livro(self, titulo, autor, genero, quantidade):
        if not titulo.strip() or not autor.strip() or quantidade <= 0:
            print(f"âŒ Dados invÃ¡lidos para o livro '{titulo}'. Cadastro nÃ£o realizado.")
            return

        livro = {"titulo": titulo, "autor": autor, "genero": genero, "quantidade": quantidade}
        self.livros.append(livro)
        print(f"ðŸ“š Livro '{titulo}' cadastrado com sucesso!")

    # Listar livros
    def listar_livros(self):
        if not self.livros:
            print("Nenhum livro cadastrado ainda.")
            return

        print("\n--- LISTA DE LIVROS ---")
        livros_tuplas = list(map(lambda l: (l["titulo"], l["autor"], l["genero"], l["quantidade"]), self.livros))
        for tupla in livros_tuplas:
            print(f"TÃ­tulo: {tupla[0]}, Autor: {tupla[1]}, GÃªnero: {tupla[2]}, Quantidade: {tupla[3]}")

    # Buscar livro pelo tÃ­tulo
    def buscar_livro_por_titulo(self, titulo):
        resultado = list(filter(lambda l: l["titulo"].lower() == titulo.lower(), self.livros))
        if resultado:
            print("\nLivro encontrado:")
            for livro in resultado:
                tupla = (livro["titulo"], livro["autor"], livro["genero"], livro["quantidade"])
                print(f"TÃ­tulo: {tupla[0]}, Autor: {tupla[1]}, GÃªnero: {tupla[2]}, Quantidade: {tupla[3]}")
        else:
            print(f"âŒ Livro '{titulo}' nÃ£o encontrado.")

    # Gerar grÃ¡fico estilo telemetria
    def gerar_grafico_telemetria(self):
        if not self.livros:
            print("NÃ£o hÃ¡ livros cadastrados para gerar grÃ¡fico.")
            return

        # Contagem por gÃªnero
        contagem = {}
        for livro in self.livros:
            contagem[livro["genero"]] = contagem.get(livro["genero"], 0) + livro["quantidade"]

        x = list(range(1, len(contagem) + 1))
        y = list(contagem.values())
        generos = list(contagem.keys())

        plt.style.use("dark_background")
        plt.figure(figsize=(10,6))
        plt.plot(x, y, marker="o", linewidth=3, color="lime", label="Quantidade de livros")
        plt.xticks(x, generos, rotation=20)
        plt.xlabel("GÃªnero")
        plt.ylabel("Quantidade")
        plt.title("Telemetria LiterÃ¡ria - Livros por GÃªnero")

        # Valores sobre os pontos
        for i, valor in enumerate(y):
            plt.text(x[i], y[i]+0.2, str(valor), ha="center", color="white", fontsize=10)

        plt.legend()
        plt.grid(True, linestyle="--", alpha=0.5)
        plt.tight_layout()
        plt.show()


# --------- Testando ---------
if __name__ == "__main__":
    biblioteca = Biblioteca()

    # Cadastro
    biblioteca.cadastrar_livro("Dom Casmurro", "Machado de Assis", "Romance", 4)
    biblioteca.cadastrar_livro("O Hobbit", "J.R.R. Tolkien", "Fantasia", 6)
    biblioteca.cadastrar_livro("O CÃ³digo Da Vinci", "Dan Brown", "Suspense", 3)
    biblioteca.cadastrar_livro("Orgulho e Preconceito", "Jane Austen", "Romance", 5)

    # Listar e buscar
    biblioteca.listar_livros()
    biblioteca.buscar_livro_por_titulo("O Hobbit")

    # GrÃ¡fico
    biblioteca.gerar_grafico_telemetria()
```
